```python
# Get content words
def get_content_words(document, remove_stop_words=False):
    stop = set(stopwords.words('english'))
    #print stop
    content_words = []
    words = document.split()
    if remove_stop_words == False:
        return words
    #remove stop words
    for ii in range(len(words)):
        word = words[ii]
        if word not in stop:
            content_words.append(word)
    return content_words

# Compute word probability
def compute_tf(content_words):
    content_words_freq = {}
    for ii in range(len(content_words)):
        w = content_words[ii]
        content_words_freq[w] = content_words_freq.get(w, 0) + 1  
        
    content_words_count = len(content_words)
    #content_word_tf = dict((k, v / float(content_words_count)) for (k, v) in content_words_freq.items())
    content_word_tf = {}
    for (k, v) in content_words_freq.items():
        content_word_tf[k] = v/float(content_words_count)
        
    return content_word_tf

# Get content words in a sentence
def get_content_words_in_sentence(content_words, sentence):
    content_words_set = set(content_words)
    sentence = sentence.split(' ')
    content_words_in_sentence  = set([word for word in sentence if word in content_words_set])
    return content_words_in_sentence

# Compute average probability of content words in a sentence
def compute_average_probability_of_words(content_word_tf, content_words_in_sentence):
    content_words_count = len(content_words_in_sentence)
    if content_words_count > 0:
        word_freq_sum = sum([content_word_tf[w] for w in content_words_in_sentence])
        word_freq_avg = word_freq_sum / content_words_count
        return word_freq_avg
    else: 
        return 0

# Find best sentence index
def find_index_of_best_sentence(content_word_tf, sentences_as_words):
    min_possible_freq = -1
    max_value = min_possible_freq
    best_sentence_index = 0
    for i, words in enumerate(sentences_as_words):
        word_freq_avg = compute_average_probability_of_words(content_word_tf, words)
        #print i, word_freq_avg
        if word_freq_avg > max_value:
            max_value = word_freq_avg
            best_sentence_index = i
    return best_sentence_index

# Update content word probability
def update_probability(content_words_tf, words_to_update):
    for w in words_to_update:
        content_words_tf[w] *= content_words_tf[w]
    return content_words_tf

# Compute sentence ratings
def compute_ratings(document, remove_stop_words=False):
    #get content words for the doc
    content_words = get_content_words(document, remove_stop_words)
    content_words_tf = compute_tf(content_words)
    document = ' '.join(document.strip().split('\n'))
    tokenizer = PunktSentenceTokenizer()
    sentences = tokenizer.tokenize(document)
    sentences_list = list(sentences)
    sentences_as_words = [get_content_words_in_sentence(content_words, s) for s in sentences]
    ratings = {}
    while len(sentences_list) > 0:
        best_sentence_index = find_index_of_best_sentence(content_words_tf, sentences_as_words)
        #print best_sentence_index
        #print sentences[best_sentence_index]
        best_sentence = sentences_list.pop(best_sentence_index)
        #print best_sentence
        # value is the iteration in which it was removed multiplied by -1 so that the first sentences removed (the most important) have highest values
        ratings[best_sentence] = -len(ratings)

        # update probabilities
        best_sentence_words = sentences_as_words.pop(best_sentence_index)
        content_words_tf = update_probability(content_words_tf, best_sentence_words)

    return ratings

    
```
